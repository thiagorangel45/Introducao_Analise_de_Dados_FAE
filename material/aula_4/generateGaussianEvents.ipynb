```python
# Importar ROOT e configurações
import ROOT
%jsroot on
```

```python
# Gerar um conjunto de dado de 1000 eventos em x a partir de uma distribuição gaussiana
# O gauss.generate() é um método da classe RooGaussian do ROOT que gera um conjunto de dados (dataset) seguindo uma distribuição gaussiana.
# O RooArgSet(x) é usado para especificar quais variáveis devem ser geradas no dataset. Neste caso, estamos gerando dados apenas na variável x.
# O 10000 é o número de eventos que queremos gerar para o dataset.
```

```python
# Definir a variável x com intervalo de -10 a 10
x = ROOT.RooRealVar("x", "x", -10, 10)

mean = ROOT.RooRealVar("mean", "Mean of Gaussian", 0, -10, 10)
width = ROOT.RooRealVar("width", "Width of Gaussian", 2, 0.1, 10)
gaussian = ROOT.RooGaussian("gaussian", "Gaussian PDF", x, mean, width)
```

```python
# Gerar 10.000 eventos a partir da PDF
data = gaussian.generate(ROOT.RooArgSet(x), 10000)
```

```python
gaussian.fitTo(data)
```

```python
frame = x.frame()

data.plotOn(frame)

# Plotar a função PDF Gaussiana ajustada
gaussian.plotOn(frame)
```

```python
# Criar um canvas para visualização
c1 = ROOT.TCanvas("c1", "Canvas", 800, 600)

frame.Draw()

c1.Draw()

#c1.SaveAs("toymc.png")
```
