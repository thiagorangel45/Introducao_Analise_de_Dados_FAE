```python
import ROOT
%jsroot on
from ROOT import RooFit
```

```python
file = ROOT.TFile.Open("/opendata/example_data.root")
tree = file.Get("nt")
```

```python
# Definir a variável observável com o nome 'mass'
x = ROOT.RooRealVar("mass", "mass", 0.0, 2.0)
```

```python
dataUnbinned = ROOT.RooDataSet("dataUnbinned", "Unbinned data", ROOT.RooArgSet(x), RooFit.Import(tree))
```

```python
print("Dados importados:")
dataUnbinned.Print("v")
```

```python
c = ROOT.TCanvas("c", "Dado importado", 900, 700)
frame = x.frame()
dataUnbinned.plotOn(frame, RooFit.MarkerColor(ROOT.kBlue), RooFit.Name("Unbinned"))
```

```python
frame.Draw()
```

```python
c.Draw()
```

```python
c.SaveAs("ImportarDados.png")
```

```python
file.Close()
```
